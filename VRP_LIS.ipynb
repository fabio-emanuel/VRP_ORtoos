```python
from vrpnetwork import NetworkVRP
import warnings
warnings.filterwarnings('ignore')
```

```python
#inicia o objeto VRP
network2 = NetworkVRP()#
```

```python
#carrega rede = links e nos
network2.load_data("nodes_lis2.csv", "links_lis2.csv")
```

```python
# define qual eeh o no do depot
depot = 1153278
```

```python
# define quais sao os nos dos clientes
clients = [452409, 773820, 885836, 1252310, 1127475, 1251998]
```

```python
# executa o VRP
route, total_distance = network2.solve_vrp(depot, clients)
    
```

```python
# transforma a solucao no VRP em elementos da rede
path_final, distancias, customers, times, links, dist_final =  network2.trata_sol_VRP(depot,clients,route)
```

```python
# cria o SHP com as informações das paradas
network2.export_shortest_customers_to_shapefile(customers,distancias,times,links,'customs_lis.shp')
```

```python
# cria o SHP com as informações das rotas
network2.export_shortest_path_to_shapefile(path_final, 'routes_lis.shp')
```
